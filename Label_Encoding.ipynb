Label Encoding


*   Converting the labels into numeric form




for example on classification of dibates it become tuff to identity class we convert them into 0,1 so that we can decide it belongs to 0,1

```python
# importing the libraries req for this
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Dataset

```python
#loading the data from csv to dataFrame
cancer_data = pd.read_csv('/content/breast_cancer_data.csv')
```

```python
# First 5 Rows of the dataFrame
cancer_data.head()
```

```python
# finding count of diff values
# diagonsis use becuase through this we can check labels value type
cancer_data['diagnosis'].value_counts()
```

```python
# Load the Label_Encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
# appending the labels to the DataFrame
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

0 --> Benign

1 --> Malignant

```python
print(cancer_data['target'].value_counts());
```

Label Encoding of iris data

```python
#loading the dataset
iris_data = pd.read_csv('/content/iris_data.csv')
```

```python
iris_data.head()
```

```python
# count diff class and value in each class
iris_data['Species'].value_counts()
```

```python
# Loading the encoder
label_encoder_iris = LabelEncoder()
```

```python
iris_labels = label_encoder_iris.fit_transform(iris_data.Species)
```

```python
# append to iris dataset
iris_data['target'] = iris_labels
```

```python
iris_data.head()
```

```python
iris_data['target'].value_counts()
```

Iris-setosa --> 0

Iris-versicolor --> 1

Iris-virginica --> 2

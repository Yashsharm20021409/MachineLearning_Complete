This is Highly Imbalanced Dataset

0 --> Legit Transactions

1 --> Fraudulent Transactions

```python
# seprating the legit and fraudulent transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under-sampling

Build a sample dataset containing similar distribution of Legit & Fraudulent Transactionds

Number of Fraudulent Transactions --> 492
(meand jiska size small ho consider that size and create newDataset with size == smaller size among both legit/fraud

```python
# picking only 492 sample row from legit
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.size)
```

Concatenate the Two DataFrame

```python
new_dataset = pd.concat([legit_sample,fraud] , axis = 0)
```

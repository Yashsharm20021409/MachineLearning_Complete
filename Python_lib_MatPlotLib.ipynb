Matplotlib lib :- use for making plot and graph

```python
# importing library
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# Math opt using numpy lib
# genrate random number between 0->10 , 100 = > genrate 100 values
# store in linear data structure
x = np.linspace(0,10,100)
# find sin value for each x value and store in y
y = np.sin(x)
# find cos value for each x value and store in y
z = np.cos(x)
```

Plotting the data

```python
# genrate cosine wave
# here x is simply variable in which random numbers are stored
# y me store h sin value so with x it will genrate sin wave
plt.plot(x,y)
plt.show()
```

```python
# genrate cosine wave
# here x is simply variable in which random numbers are stored
# z me store h cos value so with x it will genrate cos wave
plt.plot(x,z)
plt.show()
```

```python
# adding title to x-axis & y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
# parabola print
x = np.linspace(-10,10,20)
y = x**2 # formula of parabola y = x^2
plt.plot(x,y)
plt.show()
```

```python
# what kind of line you want to plot by default it is plain line
# r = color + = sym to draw
plt.plot(x,y,'r+')
plt.show()
```

```python
x = np.linspace(-1,5,50)
# np.sin(x) == y and g = color - = symbol
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

Bar Plot

```python
# create empty plot
fig = plt.figure()
# add axes
# 0,0 = origin start pos
# 1,1 = height and width of rectangle (means mentioing area in which we have plot)
ax = fig.add_axes([0,0,1,1])
# these are two variable between which we draw plot
languages = ['English','French','Spanish','Latin','German']
# 100 people speaking eng 50 = french...
people = [100, 50, 150, 40, 70]
# assign axes to variable
ax.bar(languages,people)
# assign text to labels/axes
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

pie chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

Scatter plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

3D scatter plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
